```python
prompt_social_media = [
    "Facebook",
    "Twitter",  
    "Instagram",  
    "LinkedIn",  
    "YouTube"
]


prompt_audience = [
    "secondary school student",
    "high school student",
    "University teacher",
    "researcher",
    "business executive",
    "IT professional",
    "software developer",
    "entrepreneur",
    "journalist",
    "non-profit organization member",
    "general public",
    "engineers"
]

prompt_english_level = [
    "beginner",  
    "intermediate",  
]


prompt_tone = [
    "informal",
    "inspirational",  
    "persuasive",  
    "humorous",
    "neutral",  
    "technical",  
    "Empathetic",
    "Authoritative",
    "friendly",
    "confident",
    "Playful",
    "Personal",
    "concise",
    "relatable",
    "Captivating",
    "Enthusiastic",
    "Optimistic",
    "Respectful",
    "engaging"
]

prompt_length = [
    "long",
    "short",
    "very short"
]

prompt_title = [
    "make title from summarized article",
    "Make attention-grabbing Headline",
    "Make title up to 100 characters"
]

prompt_hashtag = [
    "no use hastags",
    "use hastags",
    "use 3 hashtags that are already existing, popular and relevant"
]

prompt_perspective = [
    "first person singular",  
    "first person plural",  
    "second person",  
    "third person objective"
]

prompt_other = [
    "excluded authors and institute information",
    "excluded authors",
    "excluded institute information",
    "Add authors in the form of mentions"
]

prompt_emoji = [
    "use emoticons",
    "do not use emoticons"
]

sumarization = [
    "Begin with Thought-Provoking Question",
    "Provide the important information in the middle"
    "Finish with call to action” OR “Finish with a Poll based on the paper"
]

prompt_readability= [
    "Use words with one or two syllables where possible",
    "Use sentences with one or no conjunctions",
    "Use popular terms",
    "Do not use passive voice"
]
```

```python
import random


def generate_question():
    audience = random.choice(prompt_audience)
    english_level = random.choice(prompt_english_level)
    tone = random.choice(prompt_tone)
    length = random.choice(prompt_length)
    social_media = "LinkedIn" #random.choice(prompt_social_media)
    title_option = random.choice(prompt_title)
    emoji_option = random.choice(prompt_emoji)
    hashtag_option = random.choice(prompt_hashtag)
    perspective = random.choice(prompt_perspective)
    sum = random.choice(sumarization)
    #other = random.choice(others)
    
    question = (f"Can you create a {social_media} post: "
                f"summarize the text: {{text}}; "
                "In the post use these parameters:"
                f"use {tone} tone, "
                f"use {english_level} English, "
                f"use {length} length of post,"
                f"{emoji_option},"
                f"targeted at {audience}, "
                f"{hashtag_option}, "
                f"written from a {perspective} perspective",
                "use words with one or two syllables where possible",
                "use sentences with one or no conjunctions",
                "use popular terms",
                "do not use passive voice",
                "begin with Thought-Provoking Question"


    )
            
    
    return question

generate_question()
```

```python
questions = [generate_question() for _ in range(30)]
df_questions = pd.DataFrame({'Prompt': questions})

# Save to CSV
df_questions.to_csv('csv/NewPromptGenerated.csv', index=False)
```

```python
additional_questions = [generate_question() for _ in range(10)]

# Create a DataFrame for the new prompts
df_additional_questions = pd.DataFrame({'Prompt': additional_questions})

# Append these new prompts to the existing CSV file
with open('csv/NewPromptGenerated.csv', 'a') as f:
    df_additional_questions.to_csv(f, header=False, index=False)
```
